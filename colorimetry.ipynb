```python
import cv2
import matplotlib.pyplot as plt
```

```python
pepper = cv2.imread('pic/pepper.bmp')
pepper = cv2.cvtColor(pepper, cv2.COLOR_BGR2RGB) 
pepper = cv2.imread('pic/pepper.bmp')
pepper = cv2.cvtColor(pepper, cv2.COLOR_BGR2RGB) 
```

```python
red, green, blue = cv2.split(pepper)
fig, axs = plt.subplots(3, 4, figsize=(12, 10))
axs[0][0].imshow(pepper)
axs[0][0].set_title('Original')
axs[0][1].imshow(red, cmap='gray')
axs[0][1].set_title('Red')
axs[0][2].imshow(green, cmap='gray')
axs[0][2].set_title('Green')
axs[0][3].imshow(blue, cmap='gray')
axs[0][3].set_title('Blue')

pepper_ycbcr = cv2.cvtColor(pepper, cv2.COLOR_RGB2YCrCb)
Y, Cr, Cb = cv2.split(pepper_ycbcr)
axs[1][0].imshow(pepper)
axs[1][0].set_title('Original')
axs[1][1].imshow(Y, cmap='gray')
axs[1][1].set_title('Y')
axs[1][2].imshow(Cr, cmap='gray')
axs[1][2].set_title('Cr')
axs[1][3].imshow(Cb, cmap='gray')
axs[1][3].set_title('Cb')

pepper_HSV = cv2.cvtColor(pepper, cv2.COLOR_RGB2HSV)
H, S, I = cv2.split(pepper_HSV)
axs[2][0].imshow(pepper)
axs[2][0].set_title('Original')
axs[2][1].imshow(H, cmap='gray')
axs[2][1].set_title('Hue')
axs[2][2].imshow(S, cmap='gray')
axs[2][2].set_title('Saturation')
axs[2][3].imshow(I, cmap='gray')
axs[2][3].set_title('Intensity')
```
