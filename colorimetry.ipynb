```python
import cv2
import matplotlib.pyplot as plt
```

```python
lenna = cv2.imread('pic/lenna.png')
lenna = cv2.cvtColor(lenna, cv2.COLOR_BGR2RGB)  # OpenCV는 BGR을 사용하므로 RGB로 변환
```

```python
blue = lenna[:, :, 0]
green = lenna[:, :, 1]
red = lenna[:, :, 2]


fig, axs = plt.subplots(1, 4, figsize=(12, 4))
axs[0].imshow(lenna)
axs[0].set_title('Original')
axs[1].imshow(blue, cmap='Blues')
axs[1].set_title('Blue')
axs[2].imshow(green, cmap='Greens')
axs[2].set_title('Green')
axs[3].imshow(red, cmap='Reds')
axs[3].set_title('Red')
```

```python
lenna_HSV = cv2.cvtColor(lenna, cv2.COLOR_RGB2HSV)
H, S, I = cv2.split(lenna_HSV)

fig, axs = plt.subplots(1, 4, figsize=(12, 4))
axs[0].imshow(lenna)
axs[0].set_title('Original')
axs[1].imshow(H, cmap='hsv')
axs[1].set_title('Hue')
axs[2].imshow(S, cmap='gray')
axs[2].set_title('Saturation')
axs[3].imshow(I, cmap='gray')
axs[3].set_title('Intensity')
```

```python
lenna_ycbcr = cv2.cvtColor(lenna, cv2.COLOR_RGB2YCrCb)
Y, Cr, Cb = cv2.split(lenna_ycbcr)

fig, axs = plt.subplots(1, 4, figsize=(12, 4))
axs[0].imshow(lenna)
axs[0].set_title('Original')
axs[1].imshow(Y, cmap='gray')
axs[1].set_title('Y')
axs[2].imshow(Cr, cmap='Reds')
axs[2].set_title('Cr')
axs[3].imshow(Cb, cmap='Blues')
axs[3].set_title('Cb')
```

```python
lenna_yub = cv2.cvtColor(lenna, cv2.COLOR_RGB2YUV)
Y, U, V = cv2.split(lenna_yub)

fig, axs = plt.subplots(1, 4, figsize=(12, 4))
axs[0].imshow(lenna)
axs[0].set_title('Original')
axs[1].imshow(Y, cmap='gray')
axs[1].set_title('Y')
axs[2].imshow(U, cmap='Blues')
axs[2].set_title('U')
axs[3].imshow(V, cmap='Reds')
axs[3].set_title('V')
```
