```python
import numpy as np
import matplotlib.pyplot as plt

steps = 6
step_width = 50
height = 100
width = steps * step_width

image = np.zeros((height, width), dtype=np.uint8)
for i in range(steps):
    gray_val = (i + 1) * 255 / steps
    image[:, i * step_width:(i + 1) * step_width] = gray_val

luminance = image[0]
kernel = np.array([-0.2, 1, -0.2]) 
brightness = np.convolve(luminance, kernel, mode='valid')

fig, axs = plt.subplots(1, 2, figsize=(10, 4))
axs[0].imshow(image, cmap='gray', vmin=0, vmax=255)
axs[0].set_title("Step Brightness Image")
axs[0].axis('off')

axs[1].plot(luminance)
axs[1].set_title("Brightness")
axs[1].set_xlabel("Position")
axs[1].set_ylabel("Brightness")
axs[1].plot(brightness, color='red')

plt.tight_layout()
plt.show()
x
```
